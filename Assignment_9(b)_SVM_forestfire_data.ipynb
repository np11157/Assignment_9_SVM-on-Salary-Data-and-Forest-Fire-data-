```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import numpy as np
```

## Import Data

```python
forestfire_data = pd.read_csv('forestfires.csv')
forestfire_data
```

```python
forestfire_data.isna().sum()
```

```python
forestfire_data_1 = forestfire_data.drop(['month','day'],axis=1)
forestfire_data_1
```

```python
# Label Encoding
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
forestfire_data_1['size_category'] = le.fit_transform(forestfire_data_1['size_category'])
forestfire_data_1
```

```python
# Normalization
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return x

forestfire_data_1 = norm_func(forestfire_data_1.iloc[:,0:])
forestfire_data_1
```

```python
X = forestfire_data_1.drop('size_category',axis=1)
y = forestfire_data_1[['size_category']]
# X = forestfire_data_1.loc[:,['RH','wind','rain','area','dayfri','monthfeb','monthjan','monthjul','monthjun','monthmar','monthmay','monthnov','monthoct','monthsep']]
# y = forestfire_data_1.loc[:,['size_category']]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=12)
```

## Model Building

```python
import warnings
warnings.filterwarnings('ignore')

from sklearn.svm import SVC
s = SVC(kernel='rbf')
classifier = s.fit(x_train,y_train)
```

```python
# Model Testing
y_pred_rbf = classifier.predict(x_test)
```

```python
# Model Evalution
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix
print('Accuracy Score    : ', accuracy_score(y_test,y_pred_rbf))
print('precision Score   : ', precision_score(y_test,y_pred_rbf))
print('Recall Score      : ',recall_score(y_test,y_pred_rbf))
print('Confussion matrix : \n',confusion_matrix(y_test,y_pred_rbf))
```
